```python
#QUANTUM ENTANGLEMENT
from qiskit import QuantumCircuit , transpile
from qiskit_aer import Aer
import matplotlib.pyplot as plt

#defining a quantum circuit with 2 classical and 2 quantum registers

qc=QuantumCircuit(2,2)

qc.h(0)#appling the hadamard gate to the first qubit
qc.cx(0,1)#applying controlled not gate to the second qubit(control qubit,target qubit)
#measure the qubit 0 and store the result in the classical register 0
qc.measure(1,0)

qc.draw("mpl")
```

```python
simulator =Aer.get_backend("qasm_simulator")
circ=transpile(qc,simulator)

job=simulator.run(circ,shots=1000)

res=job.result()

counts=res.get_counts()

print(counts)

plt.bar(counts.keys(),counts.values())

plt.title("QUANTUM CIRCUIT MEASUREMENT RESULTS")
plt.xlabel("outcome")
plt.ylabel("frequency")

plt.show()
```
