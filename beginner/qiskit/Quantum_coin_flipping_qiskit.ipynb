```python
#Quantum coin flipping using qiskit
#importing all the necessary libraries
from qiskit import QuantumCircuit , transpile
from qiskit_aer import Aer
import matplotlib.pyplot as plt

#defining a quantum circuit with one quatum register (qubit) and one classical register

qc=QuantumCircuit(1,1)

qc.h(0) #applying a hadamard gate to a qubit which creates a superposition state

qc.measure(0,0) #measuring the qubits

qc.draw("mpl")



```

```python
simulator=Aer.get_backend("qasm_simulator")

circ=transpile(qc,simulator)

job=simulator.run(circ,shots=2040)

res=job.result()

counts=res.get_counts()

print(counts)

plt.bar(counts.keys(),counts.values())

plt.title("QUANTUM CIRCUIT MEASURMENT RESULTS")
plt.xlabel("outcome")
plt.ylabel("frequency(probability of occurance)")

plt.show()
```
