```python
import cirq

# create two qubits
q0=cirq.LineQubit(0)
q1=cirq.LineQubit(1)

# cretae a circuit
circuit=cirq.Circuit(
    cirq.H(q0),#applying hadamard gate to qo which crates a superposition of equal probability
    cirq.CNOT(q0,q1),# apllying the cnot gate  with q0 as control and q1 as target qubit
    cirq.measure(q0,q1,key="result")#measure both qubits stores the result under the name result
)

#simulate the circuit

simulator = cirq.Simulator()
result= simulator.run(circuit, repetitions=1000)

#print results

print(result.histogram(key="result"))

print(circuit)
```

#What does the code do?

1. Initializes two qubits
2. apply hadamard gate to first qubit to create a superposition
3. apply CNOT gate to entangle the first qubit with the second one
4. measure both qubits 1000 times to measure the entanglement
5. prints the gistogram measuremnt reults and displays the circuit
